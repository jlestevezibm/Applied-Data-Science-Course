```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import geocoder
import folium 
```

<h3> Search all pet groomer in vigo troung "Paxinas Galegas" a local telephone guide </h3>

```python
req1 = requests.get("https://www.paxinasgalegas.es/resultados.aspx?tipo=2&ncom=&aynom=vigo&cllnom=&epinom=peluquer%C3%ADas%20caninas&epi=407&ay=310&prov=&pagina=0")
req2 = requests.get("https://www.paxinasgalegas.es/resultados.aspx?tipo=2&ncom=&aynom=vigo&cllnom=&epinom=peluquer%C3%ADas%20caninas&epi=407&ay=310&prov=&pagina=1")
```

```python
soup1 = BeautifulSoup(req1.text, "xml") 
soup2 = BeautifulSoup(req2.text, "xml") 
```

```python
df = pd.DataFrame(columns = ['Name','WebPage',"Tel","Mvl","Street", "CP","LAT", "LNG"])
```

<h3> Extract all data searching in the "soup" </h3>

```python
result1 = soup1.find_all('li',attrs={'class': lambda e: e.startswith('modulo_basico') if e else False})
result2 = soup2.find_all('li',attrs={'class': lambda e: e.startswith('modulo_basico') if e else False})
result = result1 + result2

for li in result:
    fila = []
    fila.append(li.get("data-name"))
    fila.append(li.get("data-empuri"))
    fila.append(li.get("data-telf"))
    fila.append(li.get("data-whatsapp"))
    fila.append(li.find("span", {"class": "calle"}).text[:-7])
    fila.append(li.find("span", {"class": "calle"}).text[-5:])
    fila.append(li.get("data-lat"))
    fila.append(li.get("data-lng"))
    df.loc[len(df)] = fila

   
```

```python
df
```

<h3>we position all pet groomer on a map of vigo</h3>

```python
mapa = folium.Map(location = [42.223363, -8.719454], zoom_start = 12)
```

```python
for lat, lng, name  in zip(df["LAT"], df["LNG"], df["Name"]):
    folium.Marker(
        [lat, lng],
        tooltip=name,
        color='blue',).add_to(mapa) 
```

<h3>Search all postal codes of Vigo:</h3>

```python
req3 = requests.get("http://distritopostal.es/pontevedra/vigo")
```

```python
soup3 = BeautifulSoup(req3.text, "xml") 
```

```python
df2 = pd.DataFrame(columns = ["CP"])
```

```python
result3 = soup3.find_all('div',attrs={'class': lambda e: e.startswith('infocity') if e else False})

for div in result3:
    h2 = div.find_all("h2")
    for element in h2:
        if len(element.text)== 5:
            df2.loc[len(df2)] = element.text
```

```python
df2
```

<h3> Search unique values in dataframe: </h3> 

```python
dfonlycp = df[['CP']]
```

```python
dfwith=dfonlycp.groupby('CP').count()
```

<h3> Find CP Values who do not have a pet groomer: </h3>

```python
dfvoid=df2[~df2.isin(dfwith)].dropna()
```

```python
dfvoid
```

<h3> We geolocate postal codes that do not have pet groomer </h3>

```python
dfvoidwgeo = pd.DataFrame(columns = ['CP',"LAT", "LNG"])
```

```python
for index, row in dfvoid.iterrows():
    fila =[]
    fila.append(row["CP"])
    loc = geocoder.bing(row["CP"] + ",SPAIN", key="AnHrR2JwxkQorm7TSNGE6hYx3FitZw5l5kvp2vwiJ2OhWgyovQhtUinJU50_ODZb")
    fila.append(loc.lat)
    fila.append(loc.lng)
    dfvoidwgeo.loc[len(dfvoidwgeo)] = fila
```

```python
for lat, lng in zip(dfvoidwgeo["LAT"],dfvoidwgeo["LNG"]):
    folium.Marker(
        [lat, lng],
        icon=folium.Icon(color='red', icon='info-sign')
        ).add_to(mapa) 
```
